```python
# 1. Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Load Dataset
df = pd.read_csv("iris.csv")  
```

```python
df.head()
```

## üìÅDataset Description

The Iris dataset contains 150 samples of iris flowers, categorized into three species:
- Setosa  
- Versicolor  
- Virginica  

Each sample includes the following features:
- Sepal Length (cm)  
- Sepal Width (cm)  
- Petal Length (cm)  
- Petal Width (cm)

The goal is to explore the relationships between these features and the species type.

```python
# 3. Dataset Overview
print("Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nFirst 5 Rows:\n", df.head())
print("\nInfo:\n")
df.info()
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# 4. Basic Summary Statistics
print("\nSummary Statistics:\n", df.describe())
```

## üìä Exploratory Data Analysis (EDA)

We now explore the dataset visually to understand:
- How features are distributed  
- Relationships between variables  
- Outliers or unusual values using scatter plots, histograms, and box plots

```python
# 5. Scatter Plot: Petal Length vs Petal Width

# Drop ID column if it exists
if 'Id' in df.columns:
    df.drop(columns=['Id'], inplace=True)

# Rename columns to standard format (if needed)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='Set2')

plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.tight_layout()
plt.show()
```

```python
# 6. Histogram: Sepal Width Distribution
plt.figure(figsize=(7,5))
sns.histplot(data=df, x='sepal_width', kde=True, color='skyblue')
plt.title('Distribution of Sepal Width')
plt.show()
```

```python
# 7. Box Plot: Petal Width by Species
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='species', y='petal_width', hue='species', palette='pastel', legend=False)
plt.title('Petal Width by Species')
plt.show()
```

```python
# 8. Pair Plot: Full Feature Relationships
sns.pairplot(df, hue='species', palette='husl')
plt.suptitle('Pair Plot of All Features by Species', y=1.02)
plt.show()
```

```python
# 9. Heatmap: Correlation Matrix
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

## üìå Conclusion

- Sepal and petal measurements show clear differences across the three species.
- Petal length and petal width are most useful in distinguishing the species.
- Box plots and histograms reveal that Iris Setosa is the most distinct class.
- No missing values or major data quality issues were observed.
